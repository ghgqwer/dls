```python
import numpy as np
import matplotlib.pyplot as plt

# Рост (X) и вес (Y)
X = np.array([150, 160, 170, 180, 190])
Y = np.array([50, 60, 70, 80, 90])
```

```python
# Визуализация исходных данных
plt.scatter(X, Y, color='blue', label='Данные')
plt.xlabel('Рост (см)')
plt.ylabel('Вес (кг)')
plt.title('Рост и вес')
plt.legend()
plt.grid()
plt.show()
```

```python
def predict(X, m, b):
    return m * X + b
```

```python
def gradient_descent(X, Y, m, b, learning_rate, iterations):
    N = len(Y)
    
    for _ in range(iterations):
        Y_pred = predict(X, m, b)
        
        # Вычисляем градиенты
        dm = (-2/N) * np.sum(X * (Y - Y_pred))
        db = (-2/N) * np.sum(Y - Y_pred)
        
        # Обновляем параметры
        m -= learning_rate * dm
        b -= learning_rate * db
        
    return m, b
```

```python
# Начальные значения
m = 0.0
b = 0.0

# Параметры градиентного спуска
learning_rate = 0.01
iterations = 100

# Запускаем градиентный спуск
m, b = gradient_descent(X, Y, m, b, learning_rate, iterations)
m, b
```

```python
# Предсказанные значения
Y_pred = predict(X, m, b)

# Визуализация результатов
plt.scatter(X, Y, color='blue', label='Данные')
plt.plot(X, Y_pred, color='red', label='Линия регрессии')
plt.xlabel('Рост (см)')
plt.ylabel('Вес (кг)')
plt.title('Линейная регрессия: Рост и вес')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

# Пример генерации данных
np.random.seed(42)
X = np.random.rand(1000, 1)  # 1000 образцов, 1 признак
y = 3 * X.squeeze() + np.random.randn(1000) * 0.1  # y = 3x + шум
#print(f"X: {X}, y: {y}")
# Начальные веса и параметры
w = np.random.randn(1)  # случайный вес
lr = 0.01  # скорость обучения
batch_size = 100  # размер батча
n_epochs = 10  # количество эпох

for epoch in range(n_epochs):
    # Перемешиваем данные перед каждой эпохой
    indices = np.random.permutation(len(X))
    X_shuffled = X[indices]
    y_shuffled = y[indices]
    # print(f"indices = {indices}, X_shuffled = {X_shuffled}, y_shuffled = {y_shuffled}")
    # break
    for i in range(0, len(X), batch_size):
        # Формируем батч
        X_batch = X_shuffled[i:i + batch_size]
        y_batch = y_shuffled[i:i + batch_size]

        # Прогнозы (предсказания)
        y_pred = X_batch @ w

        # Вычисляем потери и градиенты
        loss = np.mean((y_pred - y_batch) ** 2)  # среднеквадратичная ошибка
        gradient = (2 / batch_size) * (X_batch.T @ (y_pred - y_batch))

        # Обновляем веса
        w -= lr * gradient

    print(f'Epoch {epoch + 1}, Loss: {loss:.4f}')

print("Обученные веса:", w)
```
