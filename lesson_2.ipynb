```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
#rng = np.random.RandomState(1)
x = 10 * np.random.rand(50)
print(x)
y = 2 * x - 5 + np.random.randn(50)
print(y)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
#Создаем модель (np.newaxis переводит одномерный массив в двумерный)
model.fit(x[:, np.newaxis], y)
#создаем тестовые данные
xfit = np.linspace(0, 10, 1000)
#Предсказываем 
yfit = model.predict(xfit[:, np.newaxis])
#строим точки
plt.scatter(x, y)
#строим график
plt.plot(xfit, yfit)
```

```python
print("Model slope:    ", model.coef_)
print("Model intercept:", model.intercept_)
```

```python
from sklearn.preprocessing import PolynomialFeatures
x = np.array([2, 3, 4])
#Создаем экземпляр класса полиномия_фьючерс
# Здесь 3 указывает, что мы хотим получить базисные функции до третьего порядка (то есть ( x^1, x^2, x^3 )). Аргумент include_bias=False означает, что мы не хотим добавлять константный член (то есть ( x^0 )).
poly = PolynomialFeatures(3, include_bias=False)
poly.fit_transform(x[:, None])
```
